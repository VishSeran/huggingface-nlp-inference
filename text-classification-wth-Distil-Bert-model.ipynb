# **Text Classification with DISTILBERT Model**

### **Import Libraries**

```python
from transformers import DistilBertForSequenceClassification, DistilBertTokenizer
import torch
```

```python
# load the model and tokenizer
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

#### **Pre-Process the data**

```python
inpu_text = "Congratulations! You've won a free ticket to the Bahamas. Reply WIN to claim."
input_tokens = tokenizer(inpu_text,return_tensors='pt')
input_tokens
```

#### **Output from the Model**


```python
# the model already trained and have optimized gradients. So to make predictions from
# the model we dont use gradient again.

with torch.no_grad():
    output = model(**input_tokens)
print(f"model output: {output}")
```

```python
# post-process output

logits = output.logits
logits, logits.shape
```

```python
probabilities = torch.softmax(logits, dim=-1)
predicted_class = torch.argmax(probabilities,dim=-1)
predicted_class
```

```python
label_class = ['negative', 'positive']
classify_class = label_class[predicted_class]
print(f"classify class label: {classify_class}")
```
