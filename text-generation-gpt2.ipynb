# **Text Generate from GPT-2 Pre-Define Model**

#### **Import Libraries**

```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
# initiate the model and the tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
## initiate the prompt
prompt = "Once a upon a time"
tokens = tokenizer(prompt, return_tensors='pt')
tokens
```

```python
#output from the model
generate_tokens = model.generate(
    tokens.input_ids,
    attention_mask = tokens.attention_mask,
    pad_token_id = tokenizer.eos_token_id,
    max_length = 100,
    num_return_sequences = 1
)
```

```python
generate_tokens
```

### **Convert Tokens to Texts**

```python
generate_texts = tokenizer.decode(generate_tokens[0], skip_special_tokens=True)
generate_texts
```

```python
with torch.no_grad():
    output = model(**tokens)
output
```

```python
output_text = tokenizer.decode(output[0], skip_special_tokens=True)
output_text
```
