# **Hugging Face Pipeleines**

The `pipeline()` function from the Hugging Face `transformers` library is a high-level interface that makes it easy to use pretrained models for different natural language processing (NLP) tasks. It hides the underlying complexity of loading models, tokenizing inputs, running inference, and formatting outputs, allowing users to perform advanced NLP tasks with minimal code.

---

## Definition

```python
transformers.pipeline(
    task: str,
    model: Optional = None,
    config: Optional = None,
    tokenizer: Optional = None,
    feature_extractor: Optional = None,
    framework: Optional = None,
    revision: str = 'main',
    use_fast: bool = True,
    model_kwargs: Dict[str, Any] = None,
    **kwargs
)

Parameters

1. task: str
* Specifies the type of task to execute, such as text classification, text generation, or question answering.
Example: "text-classification"

2. model: Optional
* Defines which model to use. This can be a model name from the Hugging Face hub, a local directory, or an already loaded model.
Example: "distilbert-base-uncased-finetuned-sst-2-english"

3. config: Optional
* Provides configuration settings for the model. It can be a name, file path, or configuration object.
Example: {"output_attentions": True}

4. tokenizer: Optional
* Specifies the tokenizer for processing input text. It can be a pretrained tokenizer name, local path, or object instance.
Example: "bert-base-uncased"

5. feature_extractor: Optional
* Used in tasks like image or audio processing to extract features from raw input data.
Example: "facebook/detectron2"

6. framework: Optional
* Indicates whether to use "pt" (PyTorch) or "tf" (TensorFlow). If not provided, it is automatically detected.

7. revision: str (default: 'main')
* Refers to a specific version of the model, such as a branch, tag, or commit.

8. use_fast: bool (default: True)
* Determines whether to use a faster tokenizer implementation when available.

9. model_kwargs: Dict[str, Any] (default: None)
* Additional arguments passed to the model during initialization.
Example: {"output_hidden_states": True}

10. kwargs: Any
* Extra arguments forwarded to different components within the pipeline.

Supported Task Types
1. Text Classification (text-classification)
Goal: Assign predefined labels to text
Examples: Sentiment analysis, spam detection, topic categorization

2. Text Generation (text-generation)
Goal: Generate meaningful text from a prompt
Examples: Story writing, chatbot responses, content creation

3. Question Answering (question-answering)
Goal: Extract answers from a given context
Examples: QA systems, document-based queries

4. Named Entity Recognition (ner / token-classification)
Goal: Detect and label entities such as names, places, or organizations
Examples: Information extraction from text

5. Summarization (summarization)
Goal: Convert long text into shorter summaries
Examples: News summarization, document compression

6. Translation (translation_xx_to_yy)
Goal: Translate text between languages
Examples: English to French translation

7. Fill-Mask (fill-mask)
Goal: Predict missing words in a sentence
Examples: Masked language modeling tasks

8. Zero-Shot Classification (zero-shot-classification)
Goal: Classify text into unseen categories without prior training
Examples: Flexible classification scenarios

9. Feature Extraction (feature-extraction)
Goal: Obtain numerical representations (embeddings) of text
Examples: Similarity search, clustering, downstream ML tasks
